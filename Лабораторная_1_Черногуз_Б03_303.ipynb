```python
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.pyplot import figure

with open('episod 2', 'r') as f:
    data = f.read().split('\n')
data = [data[i].split() for i in range(len(data))]
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=[7, 7])
fig.tight_layout()

x, y = [], []
x.append([float(data[0][i]) for i in range(len(data[0]))]) #0
y.append([float(data[1][i]) for i in range(len(data[1]))]) #0
x.append([float(data[2][i]) for i in range(len(data[2]))]) #1
y.append([float(data[3][i]) for i in range(len(data[3]))]) #1
x.append([float(data[4][i]) for i in range(len(data[4]))]) #2
y.append([float(data[5][i]) for i in range(len(data[5]))]) #2
x.append([float(data[6][i]) for i in range(len(data[6]))]) #3
y.append([float(data[7][i]) for i in range(len(data[7]))]) #3
x.append([float(data[8][i]) for i in range(len(data[8]))]) #4
y.append([float(data[9][i]) for i in range(len(data[9]))]) #4
x.append([float(data[10][i]) for i in range(len(data[10]))]) #5
y.append([float(data[11][i]) for i in range(len(data[11]))]) #5

k = 0
for i in range(3):
    for j in range (2):
        axs[i][j].plot(x[k], y[k])
        axs[i][j].xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=1))
        axs[i][j].yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=2.5))
        axs[i][j].grid(which='major', alpha=1)
        axs[i][j].set_xlim(0, 16)
        axs[i][j].set_ylim(-10, 12)
        axs[i][j].set_title("frame {id}".format(id=k))
        k += 1

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

with open('students.csv', 'r') as f:
    data = f.read().split('\n')
data = [data[i].split(';') for i in range(len(data))]
# 0 - prep, 1 - group, 2 - mark
marks_per_prep = [[], [], [], [], [], [], [], []]
#массив оценок по преподавателям
k = [[[], [], [], [], [], [], [], []], [[], [], [], [], [], [], [], []], [[], [], [], [], [], [], [], []], [[], [], [], [], [], [], [], []],
     [[], [], [], [], [], [], [], []], [[], [], [], [], [], [], [], []], [[], [], [], [], [], [], [], []], [[], [], [], [], [], [], [], []]]

marks_per_prep[1] = [data[i][2] if data[i][0] == 'prep1' else 0 for i in range(len(data))]
marks_per_prep[2] = [data[i][2] if data[i][0] == 'prep2' else 0 for i in range(len(data))]
marks_per_prep[3] = [data[i][2] if data[i][0] == 'prep3' else 0 for i in range(len(data))]
marks_per_prep[4] = [data[i][2] if data[i][0] == 'prep4' else 0 for i in range(len(data))]
marks_per_prep[5] = [data[i][2] if data[i][0] == 'prep5' else 0 for i in range(len(data))]
marks_per_prep[6] = [data[i][2] if data[i][0] == 'prep5' else 0 for i in range(len(data))]
marks_per_prep[7] = [data[i][2] if data[i][0] == 'prep5' else 0 for i in range(len(data))]

for j in range(1, 8):
  for r in range(8):
    k[r][j] = sum([int(marks_per_prep[j][i]) if marks_per_prep[j][i] == str(r) else 0 for i in range(len(marks_per_prep[1]))])

category_names = ['3', '4', '5', '6', '7', '8', '9', '10']
results = {
    'Prep 1': [k[0][1], k[1][1], k[2][1], k[3][1], k[4][1], k[5][1], k[6][1], k[7][1]],
    'Prep 2': [k[0][2], k[1][2], k[2][2], k[3][2], k[4][2], k[5][2], k[6][2], k[7][2]],
    'Prep 3': [k[0][3], k[1][3], k[2][3], k[3][3], k[4][3], k[5][3], k[6][3], k[7][3]],
    'Prep 4': [k[0][4], k[1][4], k[2][4], k[3][4], k[4][4], k[5][4], k[6][4], k[7][4]],
    'Prep 5': [k[0][5], k[1][5], k[2][5], k[3][5], k[4][5], k[5][5], k[6][5], k[7][5]],
    'Prep 6': [k[0][6], k[1][6], k[2][6], k[3][6], k[4][6], k[5][6], k[6][6], k[7][6]],
    'Prep 7': [k[0][7], k[1][7], k[2][7], k[3][7], k[4][7], k[5][7], k[6][7], k[7][7]],
}

def survey(results, category_names):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('RdYlGn')(
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        ax.barh(labels, widths, left=starts, height=0.5,
                label=colname, color=color)
        xcenters = starts + widths / 2

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(int(c)), ha='center', va='center',
                    color=text_color)
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')
    return fig, ax
survey(results, category_names)
plt.title("Marks per prep \n \n")
plt.show()
```
